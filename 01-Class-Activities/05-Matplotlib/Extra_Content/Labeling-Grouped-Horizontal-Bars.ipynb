```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random
```

```python
y1 = [random.randint(-5,10) for n in range(10)]
y2 = [random.randint(1,10) for n in range(10)]
```

```python
df = pd.DataFrame({'y1':y1, 'y2':y2})
```

```python
def set_horizontal_labels(rects):
    for rect in rects:
        x_value = rect.get_width()
        y_value = rect.get_y() + rect.get_height() / 2

        space = 5

        ha = 'left'

        if x_value < 0:
            space *= -1
            ha = 'right'

        plt.annotate(
            x_value,                    
            (x_value, y_value),         
            xytext=(space, 0),          
            textcoords="offset points", 
            va='center',                
            ha=ha)                      
```

```python
df_plot = df.plot(kind='barh')

rects = df_plot.patches

fig = plt.gcf()
fig.set_size_inches(6, 10)

set_horizontal_labels(rects)

plt.tight_layout()
```
